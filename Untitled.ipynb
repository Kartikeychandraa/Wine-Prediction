```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv("winedata.csv")
```

```python
wine.isnull().sum() # check null values
```

```python
wine.describe(include="all") #visual on all 
```

```python
import matplotlib.pyplot as plt
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
wine.hist(figsize=(15,20))
```

```python
def plot(feature_x,target='quality'):
    sns.factorplot(x=target,y=feature_x,data=wine,kind='bar',size=5,aspect=1)
    sns.factorplot(x=target,y=feature_x,data=wine,kind='violin',size=5,aspect=1)
    sns.factorplot(x=target,y=feature_x,data=wine,kind='swarm',size=5,aspect=1)
```

```python
plot('fixed acidity','quality')
```
