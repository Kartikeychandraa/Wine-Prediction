```python
wine = pd.read_csv("winedata.csv")
```

```python
import matplotlib.pyplot as plt
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
wine.hist(figsize=(15,20))
```

```python
def plot(feature_x,target='quality'):
    sns.catplot(x=target,y=feature_x,data=wine,kind='bar',height=5,aspect=1)
    sns.catplot(x=target,y=feature_x,data=wine,kind='violin',height=5,aspect=1)
    sns.catplot(x=target,y=feature_x,data=wine,kind='swarm',height=5,aspect=1)
```

```python
plot('fixed.acidity','quality')
```

```python
plot('alcohol','quality')
```
