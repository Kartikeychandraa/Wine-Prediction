```python
wine = pd.read_csv("winedata.csv")
wine.head()
#Since id is not needed
#here we will not use id 
```

```python
#id is not needed
wine=wine.drop(["id"],axis = 1)
wine.isnull().sum() 
# check null values
```

```python
import matplotlib.pyplot as plt
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
wine.hist(figsize=(15,20))
```

```python
n=list(wine.columns.values)
def plot(feature_x,target='quality'):
    plt.figure(figsize=(20,5))
    sns.boxplot(x=target,y=feature_x,data=wine)
for _ in n:
    plot(_,'quality')
```

<h2>Modeling</h2>

```python
#x=wine.iloc[:,11].values
wine.head()
```

```python
x=wine[["fixed.acidity","volatile.acidity","citric.acid","residual.sugar","chlorides","free.sulfur.dioxide","total.sulfur.dioxide","density","pH","sulphates","alcohol"]]
y=wine[["quality"]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
